# Data Joining

## Type of Joins
- Inner Join.
- Left Outer Join.
- Right Outer Join.
- Outer Join.

```python
import pandas as pd
```

```python
person = pd.read_csv('person.txt', header= None,names=['Key','Person','Age'])
```

```python
person
```

```python
group = pd.read_csv('group.txt',header = None,names=['Key','location'])
```

```python
group
```

## merge()
    Join the two DataFrames.
    
    pd.merge(left, right, how=’inner’, on=None, left_on=None, right_on=None)

- Arguments :
    
    left : First DataFrame.
    
    right : Second DataFrame.
    
    on : The common columns.
    
    left_on : Key columns in first DataFrame.
    
    right_on : Key columns in second DataFrame.
    
    how : Type of join left’, ’right’, ’outer’, ’inner’.

## Inner Join
- Only record from matched keys are in final result.
- Unmatched keys are removed.


    - Inner Join
        
        Group with common key
    
    
    - Inner Join
    
        Group g1


    - Inner Join
    
        Group g2

```python
person
```

```python
group
```

```python
pd.merge(person,group, how='inner')
```

## Left Outer Join
- Record from matched key.
- Unmatched keys from left table.

```python
person
```

```python
group
```

```python
pd.merge(person,group, how='left')
```

    returns a dataframe containing all the rows of the left dataframe.

```python
pd.merge(group,person, how='left')
```

## Right Outer Join
- Record from matched key.
- Unmatched keys from Right table.

```python
person
```

```python
group
```

```python
pd.merge(person, group,how='right')
```

    returns a dataframe containing all the rows of the right dataframe.

```python
pd.merge(group,person, how='right')
```

## Outer Join
• Record from matched key.
• Unmatched keys from both Dataset.

    returns a dataframe containing all the rows of the all dataframe.

```python
pd.merge(group,person, how='outer')
```

```python
data = pd.merge(group,person, how='outer')
```

```python
data1 = pd.merge(person,group, how='outer')
```
